```python
img=cv2.imread("/content/noisy.jpg")
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
img_denoised=cv2.fastNlMeansDenoising(img,h=30)
plt.imshow(img_denoised)
plt.title("Denoised Image")
plt.show()
```

```python
image = cv2.imread('/content/2.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

blur_image = cv2.GaussianBlur(image, (5, 5), 3, 3)

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True)

ax1.imshow(blur_image)
ax1.set_title('Original Image')

ax2.imshow(image)
ax2.set_title('Unblurred Image')

plt.show()
```

```python
src = cv2.imread("/content/image3.jpeg", 1)
src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
bg_removed=remove(src)
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True)

ax1.imshow(src)
ax1.set_title('Original Image')

ax2.imshow(bg_removed)
ax2.set_title('Background Removed')

plt.show()
```
